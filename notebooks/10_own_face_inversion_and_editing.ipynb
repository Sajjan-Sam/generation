```python
from pathlib import Path
import os, sys, io, json, zipfile, random, math, subprocess
from datetime import datetime

# =========================================================
# INSTALLS
# =========================================================
subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "torchmetrics[image]", "torch-fidelity", "lpips", "face-alignment", "numpy", "pandas", "pillow", "matplotlib", "tqdm"],
    check=True
)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from tqdm.auto import tqdm

import torch
from torchvision import transforms
from torchmetrics.image.fid import FrechetInceptionDistance
from torchmetrics.image.kid import KernelInceptionDistance
import lpips
import face_alignment

# =========================================================
# SETTINGS
# =========================================================
REAL_REF_COUNT = 1000
FAKE_COUNT_PER_MODEL = 1000
LPIPS_PAIR_COUNT = 200
LANDMARK_EVAL_COUNT = 300
TRUNC_PSI = 1.0
RANDOM_SEED = 42

COMPUTE_LPIPS = True
COMPUTE_LANDMARK_PLAUSIBILITY = True

# =========================================================
# PATHS
# =========================================================
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()

dataset_zip = ROOT / "data" / "stylegan3_zip" / "celeba_256x256.zip"
stylegan3_repo = ROOT / "third_party" / "stylegan3"

baseline_root = ROOT / "checkpoints" / "baseline_stylegan3r"
landmark_root = ROOT / "checkpoints" / "landmark_reg"
freq_root = ROOT / "checkpoints" / "freq_ablation"

results_root = ROOT / "results" / "metrics_eval"
results_root.mkdir(parents=True, exist_ok=True)

real_ref_dir = results_root / f"real_reference_{REAL_REF_COUNT}"
real_ref_dir.mkdir(parents=True, exist_ok=True)

fake_root = results_root / "generated_eval"
fake_root.mkdir(parents=True, exist_ok=True)

tables_dir = ROOT / "results" / "tables"
tables_dir.mkdir(parents=True, exist_ok=True)

figures_dir = ROOT / "figures" / "report"
figures_dir.mkdir(parents=True, exist_ok=True)

stats_npz = ROOT / "data" / "manifests" / "real_landmark_stats.npz"

# =========================================================
# PRECHECKS
# =========================================================
assert dataset_zip.exists(), f"Missing dataset ZIP: {dataset_zip}"
assert stylegan3_repo.exists(), f"Missing stylegan3 repo: {stylegan3_repo}"
assert stats_npz.exists(), f"Missing landmark stats NPZ: {stats_npz}"

# =========================================================
# IMPORT STYLEGAN3 PYTHON MODULES
# =========================================================
sys.path.insert(0, str(stylegan3_repo))
import dnnlib
import legacy

# =========================================================
# HELPERS
# =========================================================
IMAGE_EXTS = (".png", ".jpg", ".jpeg", ".webp", ".bmp")

def list_images(folder: Path):
    return sorted([p for p in folder.rglob("*") if p.is_file() and p.suffix.lower() in IMAGE_EXTS])

def latest_run_dir(root: Path):
    runs = [p for p in root.iterdir() if p.is_dir()] if root.exists() else []
    runs = sorted(runs, key=lambda p: p.stat().st_mtime, reverse=True)
    return runs[0] if runs else None

def latest_snapshot(run_dir: Path):
    if run_dir is None or not run_dir.exists():
        return None
    snaps = sorted(run_dir.glob("network-snapshot-*.pkl"))
    return snaps[-1] if snaps else None

def load_image_from_zip(zip_path: Path, inner_name: str):
    with zipfile.ZipFile(zip_path, "r") as zf:
        data = zf.read(inner_name)
    img = Image.open(io.BytesIO(data)).convert("RGB")
    return img

def list_images_in_zip(zip_path: Path):
    with zipfile.ZipFile(zip_path, "r") as zf:
        names = [n for n in zf.namelist() if n.lower().endswith(IMAGE_EXTS)]
    return sorted(names)

def save_rgb(img: Image.Image, path: Path):
    path.parent.mkdir(parents=True, exist_ok=True)
    img.save(path, format="PNG")

def ensure_real_reference():
    existing = list_images(real_ref_dir)
    if len(existing) >= REAL_REF_COUNT:
        print(f"[OK] Real reference already exists: {len(existing)} images")
        return

    all_zip_imgs = list_images_in_zip(dataset_zip)
    assert len(all_zip_imgs) >= REAL_REF_COUNT, "Not enough images in dataset ZIP"

    rng = random.Random(RANDOM_SEED)
    chosen = rng.sample(all_zip_imgs, REAL_REF_COUNT)

    for i, inner_name in enumerate(tqdm(chosen, desc="Extracting real reference images")):
        img = load_image_from_zip(dataset_zip, inner_name)
        out_path = real_ref_dir / f"real_{i:05d}.png"
        save_rgb(img, out_path)

def generate_fake_images(snapshot_path: Path, out_dir: Path, count: int, trunc_psi=1.0, seed_offset=0):
    out_dir.mkdir(parents=True, exist_ok=True)
    existing = list_images(out_dir)
    if len(existing) >= count:
        print(f"[OK] {out_dir.name}: already has {len(existing)} images")
        return

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Loading snapshot: {snapshot_path}")
    with dnnlib.util.open_url(str(snapshot_path)) as f:
        G = legacy.load_network_pkl(f)["G_ema"].to(device)

    label = torch.zeros([1, G.c_dim], device=device) if G.c_dim != 0 else None

    for i in tqdm(range(count), desc=f"Generating {out_dir.name}"):
        seed = seed_offset + i
        rnd = np.random.RandomState(seed)
        z = torch.from_numpy(rnd.randn(1, G.z_dim)).to(device=device, dtype=torch.float32)
        with torch.no_grad():
            img = G(z, label, truncation_psi=trunc_psi, noise_mode="const")
        img = (img * 127.5 + 128).clamp(0, 255).to(torch.uint8)
        img = img[0].permute(1, 2, 0).cpu().numpy()
        Image.fromarray(img, "RGB").save(out_dir / f"fake_{i:05d}.png")

def load_tensor_image(path: Path):
    img = Image.open(path).convert("RGB")
    x = transforms.ToTensor()(img).unsqueeze(0)
    return x

# ---------------- Landmark plausibility ----------------
npz = np.load(stats_npz, allow_pickle=True)
FEATURE_NAMES = [str(x) for x in npz["feature_names"].tolist()]
REAL_MEAN = np.asarray(npz["mean"], dtype=np.float32)
REAL_COV = np.asarray(npz["cov"], dtype=np.float32)
REAL_PREC = np.linalg.pinv(REAL_COV + np.eye(REAL_COV.shape[0], dtype=np.float32) * 1e-6)

def features_from_landmarks(lm):
    lm = np.asarray(lm, dtype=np.float32)
    left_eye = lm[36:42].mean(axis=0)
    right_eye = lm[42:48].mean(axis=0)
    nose = lm[30]
    mouth_l = lm[48]
    mouth_r = lm[54]
    mouth_center = (mouth_l + mouth_r) / 2.0
    jaw_l = lm[0]
    jaw_r = lm[16]
    chin = lm[8]
    brow_l = lm[17:22].mean(axis=0)
    brow_r = lm[22:27].mean(axis=0)
    eye_center = (left_eye + right_eye) / 2.0

    inter_ocular = np.linalg.norm(left_eye - right_eye) + 1e-8
    jaw_width = np.linalg.norm(jaw_l - jaw_r) + 1e-8
    face_height = np.linalg.norm(eye_center - chin) + 1e-8

    eye_to_nose_ratio = ((np.linalg.norm(left_eye - nose) + np.linalg.norm(right_eye - nose)) / 2.0) / inter_ocular
    mouth_width_ratio = np.linalg.norm(mouth_l - mouth_r) / inter_ocular
    nose_to_mouth_ratio = np.linalg.norm(nose - mouth_center) / inter_ocular
    left_brow_to_eye_ratio = np.linalg.norm(brow_l - left_eye) / inter_ocular
    right_brow_to_eye_ratio = np.linalg.norm(brow_r - right_eye) / inter_ocular
    eye_symmetry = abs(np.linalg.norm(left_eye - nose) - np.linalg.norm(right_eye - nose)) / inter_ocular
    brow_symmetry = abs(np.linalg.norm(brow_l - left_eye) - np.linalg.norm(brow_r - right_eye)) / inter_ocular
    jaw_symmetry = abs(np.linalg.norm(jaw_l - nose) - np.linalg.norm(jaw_r - nose)) / jaw_width
    face_hw_ratio = face_height / jaw_width

    return np.array([
        eye_to_nose_ratio,
        mouth_width_ratio,
        nose_to_mouth_ratio,
        left_brow_to_eye_ratio,
        right_brow_to_eye_ratio,
        eye_symmetry,
        brow_symmetry,
        jaw_symmetry,
        face_hw_ratio,
    ], dtype=np.float32)

def mahalanobis_batch(features, mean, prec):
    diffs = features - mean[None, :]
    d2 = np.einsum("bi,ij,bj->b", diffs, prec, diffs)
    return d2

fa_device = "cuda" if torch.cuda.is_available() else "cpu"
fa = face_alignment.FaceAlignment(
    face_alignment.LandmarksType.TWO_D,
    flip_input=False,
    device=fa_device
)

def get_landmarks_np(img_np):
    try:
        preds = fa.get_landmarks_from_image(img_np)
    except Exception:
        try:
            preds = fa.get_landmarks(img_np)
        except Exception:
            preds = None
    if preds is None or len(preds) == 0:
        return None
    lm = np.asarray(preds[0], dtype=np.float32)
    if lm.shape[0] < 68:
        return None
    return lm[:, :2]

def compute_landmark_plausibility(fake_dir: Path, max_count=300):
    imgs = list_images(fake_dir)[:max_count]
    feats = []
    failed = 0
    for p in tqdm(imgs, desc=f"Landmark plausibility {fake_dir.name}"):
        try:
            img_np = np.array(Image.open(p).convert("RGB"))
            lm = get_landmarks_np(img_np)
            if lm is None:
                failed += 1
                continue
            feats.append(features_from_landmarks(lm))
        except Exception:
            failed += 1

    if len(feats) == 0:
        return {
            "landmark_success_count": 0,
            "landmark_fail_count": failed,
            "landmark_mahalanobis_mean": np.nan,
            "landmark_mahalanobis_std": np.nan,
        }

    feats = np.stack(feats)
    d2 = mahalanobis_batch(feats, REAL_MEAN, REAL_PREC)

    return {
        "landmark_success_count": int(len(feats)),
        "landmark_fail_count": int(failed),
        "landmark_mahalanobis_mean": float(np.mean(d2)),
        "landmark_mahalanobis_std": float(np.std(d2)),
    }

# ---------------- LPIPS diversity ----------------
lpips_model = lpips.LPIPS(net="alex")
if torch.cuda.is_available():
    lpips_model = lpips_model.cuda()

def compute_lpips_diversity(fake_dir: Path, pair_count=200):
    imgs = list_images(fake_dir)
    assert len(imgs) >= 2, f"Need at least 2 images in {fake_dir}"
    rng = random.Random(RANDOM_SEED)
    pairs = []
    for _ in range(pair_count):
        a, b = rng.sample(imgs, 2)
        pairs.append((a, b))

    vals = []
    for a, b in tqdm(pairs, desc=f"LPIPS {fake_dir.name}"):
        img_a = transforms.ToTensor()(Image.open(a).convert("RGB")).unsqueeze(0) * 2 - 1
        img_b = transforms.ToTensor()(Image.open(b).convert("RGB")).unsqueeze(0) * 2 - 1
        if torch.cuda.is_available():
            img_a = img_a.cuda()
            img_b = img_b.cuda()
        with torch.no_grad():
            d = lpips_model(img_a, img_b).item()
        vals.append(d)
    return {
        "lpips_mean": float(np.mean(vals)),
        "lpips_std": float(np.std(vals)),
    }

# =========================================================
# SNAPSHOTS
# =========================================================
baseline_snapshot = latest_snapshot(latest_run_dir(baseline_root))
landmark_snapshot = latest_snapshot(latest_run_dir(landmark_root))
freq_snapshot = latest_snapshot(latest_run_dir(freq_root))

assert baseline_snapshot is not None, "Missing baseline snapshot"
assert landmark_snapshot is not None, "Missing landmark snapshot"
assert freq_snapshot is not None, "Missing frequency ablation snapshot"

model_specs = {
    "baseline": {"snapshot": baseline_snapshot, "out_dir": fake_root / "baseline"},
    "landmark_reg": {"snapshot": landmark_snapshot, "out_dir": fake_root / "landmark_reg"},
    "freq_ablation": {"snapshot": freq_snapshot, "out_dir": fake_root / "freq_ablation"},
}

print("Snapshots:")
for k, v in model_specs.items():
    print(k, "->", v["snapshot"])

# =========================================================
# PREPARE REAL + FAKE IMAGES
# =========================================================
ensure_real_reference()

seed_offsets = {
    "baseline": 0,
    "landmark_reg": 10000,
    "freq_ablation": 20000,
}

for name, spec in model_specs.items():
    generate_fake_images(
        snapshot_path=spec["snapshot"],
        out_dir=spec["out_dir"],
        count=FAKE_COUNT_PER_MODEL,
        trunc_psi=TRUNC_PSI,
        seed_offset=seed_offsets[name],
    )

# =========================================================
# COMPUTE FID + KID
# =========================================================
real_imgs = list_images(real_ref_dir)
assert len(real_imgs) >= REAL_REF_COUNT, f"Need {REAL_REF_COUNT} real images"

summary_rows = []

for model_name, spec in model_specs.items():
    fake_imgs = list_images(spec["out_dir"])
    assert len(fake_imgs) >= FAKE_COUNT_PER_MODEL, f"{model_name} has insufficient fake images"

    fid = FrechetInceptionDistance(feature=2048, normalize=True).set_dtype(torch.float64)
    kid = KernelInceptionDistance(
        feature=2048,
        subsets=10,
        subset_size=min(100, REAL_REF_COUNT, FAKE_COUNT_PER_MODEL),
        normalize=True,
    )

    for p in tqdm(real_imgs[:REAL_REF_COUNT], desc=f"FID/KID real -> {model_name}"):
        x = load_tensor_image(p)
        fid.update(x, real=True)
        kid.update(x, real=True)

    for p in tqdm(fake_imgs[:FAKE_COUNT_PER_MODEL], desc=f"FID/KID fake -> {model_name}"):
        x = load_tensor_image(p)
        fid.update(x, real=False)
        kid.update(x, real=False)

    fid_val = float(fid.compute().cpu())
    kid_mean, kid_std = kid.compute()

    row = {
        "model": model_name,
        "snapshot": str(spec["snapshot"]),
        "num_real": REAL_REF_COUNT,
        "num_fake": FAKE_COUNT_PER_MODEL,
        "fid": fid_val,
        "kid_mean": float(kid_mean.cpu()) if torch.isfinite(kid_mean) else np.nan,
        "kid_std": float(kid_std.cpu()) if torch.isfinite(kid_std) else np.nan,
    }

    if COMPUTE_LPIPS:
        row.update(compute_lpips_diversity(spec["out_dir"], pair_count=LPIPS_PAIR_COUNT))

    if COMPUTE_LANDMARK_PLAUSIBILITY:
        row.update(compute_landmark_plausibility(spec["out_dir"], max_count=LANDMARK_EVAL_COUNT))

    summary_rows.append(row)

# =========================================================
# SAVE TABLES
# =========================================================
df = pd.DataFrame(summary_rows)
csv_path = tables_dir / "metrics_summary.csv"
df.to_csv(csv_path, index=False)

json_path = tables_dir / "metrics_summary.json"
json_path.write_text(df.to_json(orient="records", indent=2), encoding="utf-8")

print("\nSaved metrics CSV:", csv_path)
print("Saved metrics JSON:", json_path)
print(df)

# =========================================================
# PLOTS
# =========================================================
plt.figure(figsize=(7, 4))
plt.bar(df["model"], df["fid"])
plt.ylabel("FID (lower is better)")
plt.title("FID comparison")
plt.tight_layout()
fid_plot = figures_dir / "metrics_fid_bar.png"
plt.savefig(fid_plot, dpi=200)
plt.show()

plt.figure(figsize=(7, 4))
plt.bar(df["model"], df["kid_mean"].fillna(0))
plt.ylabel("KID mean (lower is better)")
plt.title("KID comparison")
plt.tight_layout()
kid_plot = figures_dir / "metrics_kid_bar.png"
plt.savefig(kid_plot, dpi=200)
plt.show()

if "lpips_mean" in df.columns:
    plt.figure(figsize=(7, 4))
    plt.bar(df["model"], df["lpips_mean"])
    plt.ylabel("LPIPS diversity (higher = more diverse)")
    plt.title("LPIPS diversity comparison")
    plt.tight_layout()
    lpips_plot = figures_dir / "metrics_lpips_bar.png"
    plt.savefig(lpips_plot, dpi=200)
    plt.show()

if "landmark_mahalanobis_mean" in df.columns:
    plt.figure(figsize=(7, 4))
    plt.bar(df["model"], df["landmark_mahalanobis_mean"])
    plt.ylabel("Landmark Mahalanobis mean (lower is better)")
    plt.title("Landmark plausibility comparison")
    plt.tight_layout()
    lm_plot = figures_dir / "metrics_landmark_bar.png"
    plt.savefig(lm_plot, dpi=200)
    plt.show()

# =========================================================
# EXPORT SUMMARY
# =========================================================
summary = {
    "time": datetime.now().isoformat(),
    "dataset_zip": str(dataset_zip),
    "real_ref_dir": str(real_ref_dir),
    "fake_root": str(fake_root),
    "baseline_snapshot": str(baseline_snapshot),
    "landmark_snapshot": str(landmark_snapshot),
    "freq_snapshot": str(freq_snapshot),
    "metrics_csv": str(csv_path),
    "metrics_json": str(json_path),
    "real_ref_count": REAL_REF_COUNT,
    "fake_count_per_model": FAKE_COUNT_PER_MODEL,
    "lpips_pair_count": LPIPS_PAIR_COUNT if COMPUTE_LPIPS else 0,
    "landmark_eval_count": LANDMARK_EVAL_COUNT if COMPUTE_LANDMARK_PLAUSIBILITY else 0,
}
summary_path = tables_dir / "metrics_eval_summary.json"
summary_path.write_text(json.dumps(summary, indent=2), encoding="utf-8")

print("\nSaved summary:", summary_path)
print("Notebook 09 complete.")
print("Next: build final 15-image set and report figures.")
```

```python
from pathlib import Path
import os, sys, io, json, math, random, subprocess
from datetime import datetime

# =========================================================
# INSTALLS
# =========================================================
subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "facenet-pytorch", "lpips", "scikit-image", "numpy", "pandas", "pillow", "matplotlib", "tqdm"],
    check=True
)

import numpy as np
import pandas as pd
from PIL import Image, ImageOps, ImageEnhance, ImageFilter, ImageDraw, ImageFont
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from skimage.metrics import structural_similarity as ssim_metric
from skimage.metrics import peak_signal_noise_ratio as psnr_metric

import torch
import torch.nn.functional as F
from torchvision import transforms
import lpips
from facenet_pytorch import MTCNN, InceptionResnetV1

# =========================================================
# SETTINGS
# =========================================================
RANDOM_SEED = 42
FACE_IMAGE_NAME = "me.jpeg"     # your image placed in data/raw/personal_raw/
IMAGE_SIZE = 160                # verifier works best on 160x160 cropped faces
MTCNN_MARGIN = 20
PAIR_COUNT = 15

# If threshold is None, it is calibrated automatically from mild augmentations
MANUAL_SIMILARITY_THRESHOLD = None

# =========================================================
# PATHS
# =========================================================
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()

personal_raw_dir = ROOT / "data" / "raw" / "personal_raw"
personal_img_path = personal_raw_dir / FACE_IMAGE_NAME

personal_root = ROOT / "results" / "personal_attack_eval"
personal_root.mkdir(parents=True, exist_ok=True)

aligned_dir = personal_root / "aligned"
aligned_dir.mkdir(parents=True, exist_ok=True)

pairs_dir = personal_root / "pairs"
pairs_dir.mkdir(parents=True, exist_ok=True)

grids_dir = personal_root / "grids"
grids_dir.mkdir(parents=True, exist_ok=True)

tables_dir = personal_root / "tables"
tables_dir.mkdir(parents=True, exist_ok=True)

# =========================================================
# PRECHECKS
# =========================================================
assert personal_img_path.exists(), f"Missing personal image: {personal_img_path}"

# =========================================================
# DEVICE / MODELS
# =========================================================
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# facenet-pytorch documents MTCNN + InceptionResnetV1 for face detection and recognition.
mtcnn = MTCNN(
    image_size=IMAGE_SIZE,
    margin=MTCNN_MARGIN,
    keep_all=False,
    device=device
)
resnet = InceptionResnetV1(pretrained="vggface2").eval().to(device)

lpips_model = lpips.LPIPS(net="alex")
if torch.cuda.is_available():
    lpips_model = lpips_model.cuda()
lpips_model.eval()

# =========================================================
# HELPERS
# =========================================================
def save_img(img, path):
    path.parent.mkdir(parents=True, exist_ok=True)
    img.save(path)

def to_uint8_pil(arr):
    arr = np.clip(arr, 0, 255).astype(np.uint8)
    return Image.fromarray(arr)

def cosine_similarity(a, b):
    a = F.normalize(a, dim=1)
    b = F.normalize(b, dim=1)
    return float((a * b).sum(dim=1).item())

def pil_to_verifier_tensor(img_pil):
    # aligned crop to tensor in the format expected by InceptionResnetV1
    x = transforms.ToTensor()(img_pil).unsqueeze(0)
    # fallback normalization close to facenet-pytorch crop convention
    x = (x - 0.5) / 0.5
    return x.to(device)

def embedding_from_aligned_pil(img_pil):
    x = pil_to_verifier_tensor(img_pil)
    with torch.no_grad():
        emb = resnet(x)
    return emb

def lpips_distance(img_a_pil, img_b_pil):
    ta = transforms.ToTensor()(img_a_pil).unsqueeze(0) * 2 - 1
    tb = transforms.ToTensor()(img_b_pil).unsqueeze(0) * 2 - 1
    if torch.cuda.is_available():
        ta = ta.cuda()
        tb = tb.cuda()
    with torch.no_grad():
        d = lpips_model(ta, tb).item()
    return float(d)

def visual_metrics(img_a_pil, img_b_pil):
    a = np.array(img_a_pil.convert("RGB"))
    b = np.array(img_b_pil.convert("RGB"))
    ssim_val = ssim_metric(a, b, channel_axis=2, data_range=255)
    psnr_val = psnr_metric(a, b, data_range=255)
    lpips_val = lpips_distance(img_a_pil, img_b_pil)
    return float(ssim_val), float(psnr_val), float(lpips_val)

def draw_text(img_pil, text):
    img = img_pil.copy()
    draw = ImageDraw.Draw(img)
    try:
        font = ImageFont.load_default()
    except Exception:
        font = None
    draw.rectangle((0, 0, img.width, 18), fill=(255, 255, 255))
    draw.text((4, 2), text, fill=(0, 0, 0), font=font)
    return img

def panel_two_images(left_img, right_img, left_label, right_label, out_path):
    left = draw_text(ImageOps.pad(left_img.convert("RGB"), (IMAGE_SIZE, IMAGE_SIZE)), left_label)
    right = draw_text(ImageOps.pad(right_img.convert("RGB"), (IMAGE_SIZE, IMAGE_SIZE)), right_label)
    panel = Image.new("RGB", (IMAGE_SIZE * 2, IMAGE_SIZE), "white")
    panel.paste(left, (0, 0))
    panel.paste(right, (IMAGE_SIZE, 0))
    save_img(panel, out_path)

def make_grid(image_paths, out_path, cols=3, cell_w=320, cell_h=160, title=None):
    rows = math.ceil(len(image_paths) / cols)
    top_pad = 30 if title else 0
    canvas = Image.new("RGB", (cols * cell_w, rows * cell_h + top_pad), "white")
    draw = ImageDraw.Draw(canvas)
    try:
        font = ImageFont.load_default()
    except Exception:
        font = None
    if title:
        draw.text((10, 8), title, fill=(0, 0, 0), font=font)
    for idx, p in enumerate(image_paths):
        img = Image.open(p).convert("RGB")
        x = (idx % cols) * cell_w
        y = top_pad + (idx // cols) * cell_h
        canvas.paste(ImageOps.pad(img, (cell_w, cell_h)), (x, y))
    save_img(canvas, out_path)

# =========================================================
# LOAD AND ALIGN ORIGINAL FACE
# =========================================================
raw_img = Image.open(personal_img_path).convert("RGB")
raw_copy_path = aligned_dir / "raw_original_me.png"
save_img(raw_img, raw_copy_path)

aligned_save_path = aligned_dir / "aligned_original_me.png"
aligned_tensor = mtcnn(raw_img, save_path=str(aligned_save_path))
assert aligned_tensor is not None, (
    "MTCNN could not detect a face in me.jpeg. Use a clearer front-facing image."
)

aligned_pil = Image.open(aligned_save_path).convert("RGB").resize((IMAGE_SIZE, IMAGE_SIZE))
aligned_save_path = aligned_dir / "aligned_original_me_160.png"
save_img(aligned_pil, aligned_save_path)

orig_emb = embedding_from_aligned_pil(aligned_pil)

# =========================================================
# THRESHOLD CALIBRATION
# Attack success = perturbed image fails to match original under verifier threshold.
# If no external target model/threshold is specified, use a self-calibrated threshold.
# =========================================================
def mild_augments(img):
    out = []
    out.append(ImageEnhance.Brightness(img).enhance(0.97))
    out.append(ImageEnhance.Brightness(img).enhance(1.03))
    out.append(ImageEnhance.Contrast(img).enhance(0.97))
    out.append(ImageEnhance.Contrast(img).enhance(1.03))
    out.append(img.filter(ImageFilter.GaussianBlur(radius=0.5)))
    return out

cal_sims = []
for aug in mild_augments(aligned_pil):
    emb = embedding_from_aligned_pil(aug)
    cal_sims.append(cosine_similarity(orig_emb, emb))

cal_mean = float(np.mean(cal_sims))
cal_std = float(np.std(cal_sims))
if MANUAL_SIMILARITY_THRESHOLD is None:
    similarity_threshold = max(0.20, cal_mean - 2.0 * cal_std)
else:
    similarity_threshold = float(MANUAL_SIMILARITY_THRESHOLD)

print("Calibration similarities:", cal_sims)
print("Calibrated similarity threshold:", similarity_threshold)

# =========================================================
# PERTURBATION FUNCTIONS
# =========================================================
rng = np.random.default_rng(RANDOM_SEED)

def add_gaussian_noise(img, sigma):
    arr = np.array(img).astype(np.float32)
    noise = rng.normal(0, sigma, arr.shape)
    arr = np.clip(arr + noise, 0, 255)
    return Image.fromarray(arr.astype(np.uint8))

def jpeg_compress(img, quality):
    buf = io.BytesIO()
    img.save(buf, format="JPEG", quality=quality)
    buf.seek(0)
    return Image.open(buf).convert("RGB")

def rotate_fill(img, deg):
    return img.rotate(deg, resample=Image.Resampling.BICUBIC, expand=False, fillcolor=(128,128,128))

def downsample_restore(img, down_to):
    small = img.resize((down_to, down_to), Image.Resampling.BILINEAR)
    return small.resize((IMAGE_SIZE, IMAGE_SIZE), Image.Resampling.BILINEAR)

def occlude_rect(img, box, color=(0,0,0)):
    out = img.copy()
    draw = ImageDraw.Draw(out)
    draw.rectangle(box, fill=color)
    return out

perturbations = [
    ("noise_sigma_6", lambda x: add_gaussian_noise(x, 6)),
    ("noise_sigma_12", lambda x: add_gaussian_noise(x, 12)),
    ("blur_r1", lambda x: x.filter(ImageFilter.GaussianBlur(radius=1.0))),
    ("blur_r2", lambda x: x.filter(ImageFilter.GaussianBlur(radius=2.0))),
    ("jpeg_q80", lambda x: jpeg_compress(x, 80)),
    ("jpeg_q50", lambda x: jpeg_compress(x, 50)),
    ("brightness_0.8", lambda x: ImageEnhance.Brightness(x).enhance(0.80)),
    ("brightness_1.25", lambda x: ImageEnhance.Brightness(x).enhance(1.25)),
    ("contrast_0.8", lambda x: ImageEnhance.Contrast(x).enhance(0.80)),
    ("contrast_1.25", lambda x: ImageEnhance.Contrast(x).enhance(1.25)),
    ("rotate_p5", lambda x: rotate_fill(x, 5)),
    ("rotate_m5", lambda x: rotate_fill(x, -5)),
    ("occlude_eyes", lambda x: occlude_rect(x, (35, 40, 125, 65), color=(0,0,0))),
    ("occlude_mouth", lambda x: occlude_rect(x, (45, 105, 115, 135), color=(0,0,0))),
    ("down_up_96", lambda x: downsample_restore(x, 96)),
]

assert len(perturbations) == PAIR_COUNT, f"Expected {PAIR_COUNT} perturbations"

# =========================================================
# GENERATE 15 PAIRS
# =========================================================
rows = []
pair_panel_paths = []

for idx, (name, fn) in enumerate(perturbations, start=1):
    pair_dir = pairs_dir / f"pair_{idx:02d}_{name}"
    pair_dir.mkdir(parents=True, exist_ok=True)

    original_path = pair_dir / "original.png"
    perturbed_path = pair_dir / "perturbed.png"
    panel_path = pair_dir / "pair_panel.png"

    # Original in every pair = same aligned face
    save_img(aligned_pil, original_path)

    perturbed_img = fn(aligned_pil).convert("RGB").resize((IMAGE_SIZE, IMAGE_SIZE))
    save_img(perturbed_img, perturbed_path)

    pert_emb = embedding_from_aligned_pil(perturbed_img)
    sim = cosine_similarity(orig_emb, pert_emb)
    attack_success = int(sim < similarity_threshold)

    ssim_val, psnr_val, lpips_val = visual_metrics(aligned_pil, perturbed_img)

    panel_two_images(
        aligned_pil, perturbed_img,
        left_label="original",
        right_label=name,
        out_path=panel_path
    )
    pair_panel_paths.append(panel_path)

    rows.append({
        "pair_id": idx,
        "perturbation": name,
        "original_path": str(original_path),
        "perturbed_path": str(perturbed_path),
        "panel_path": str(panel_path),
        "cosine_similarity": sim,
        "attack_success": attack_success,
        "ssim": ssim_val,
        "psnr": psnr_val,
        "lpips": lpips_val,
        "threshold_used": similarity_threshold,
    })

df = pd.DataFrame(rows)

# =========================================================
# SAVE TABLES
# =========================================================
pairs_csv = tables_dir / "personal_attack_pairs.csv"
pairs_json = tables_dir / "personal_attack_pairs.json"
df.to_csv(pairs_csv, index=False)
pairs_json.write_text(df.to_json(orient="records", indent=2), encoding="utf-8")

attack_success_rate = float(df["attack_success"].mean())
summary = {
    "time": datetime.now().isoformat(),
    "raw_image": str(personal_img_path),
    "aligned_image": str(aligned_save_path),
    "pair_count": int(len(df)),
    "similarity_threshold": float(similarity_threshold),
    "calibration_similarity_mean": cal_mean,
    "calibration_similarity_std": cal_std,
    "attack_success_rate": attack_success_rate,
    "mean_ssim": float(df["ssim"].mean()),
    "mean_psnr": float(df["psnr"].mean()),
    "mean_lpips": float(df["lpips"].mean()),
    "pairs_csv": str(pairs_csv),
}
summary_path = tables_dir / "personal_attack_summary.json"
summary_path.write_text(json.dumps(summary, indent=2), encoding="utf-8")

print(json.dumps(summary, indent=2))
print(df[["pair_id", "perturbation", "cosine_similarity", "attack_success", "ssim", "psnr", "lpips"]])

# =========================================================
# FIGURES
# =========================================================
overview_grid = grids_dir / "personal_pairs_overview_grid.png"
make_grid(
    pair_panel_paths,
    out_path=overview_grid,
    cols=3,
    cell_w=320,
    cell_h=160,
    title="15 original–perturbed pairs for personal face"
)

# Bar charts
plt.figure(figsize=(10, 4))
plt.bar(df["perturbation"], df["cosine_similarity"])
plt.axhline(similarity_threshold, linestyle="--")
plt.xticks(rotation=70, ha="right")
plt.ylabel("Verifier cosine similarity")
plt.title("Identity similarity: original vs perturbed")
plt.tight_layout()
sim_plot = grids_dir / "personal_similarity_bar.png"
plt.savefig(sim_plot, dpi=200)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(df["perturbation"], df["ssim"])
plt.xticks(rotation=70, ha="right")
plt.ylabel("SSIM")
plt.title("Visual similarity (SSIM)")
plt.tight_layout()
ssim_plot = grids_dir / "personal_ssim_bar.png"
plt.savefig(ssim_plot, dpi=200)
plt.show()

plt.figure(figsize=(10, 4))
plt.bar(df["perturbation"], df["lpips"])
plt.xticks(rotation=70, ha="right")
plt.ylabel("LPIPS (lower = more similar)")
plt.title("Perceptual distance (LPIPS)")
plt.tight_layout()
lpips_plot = grids_dir / "personal_lpips_bar.png"
plt.savefig(lpips_plot, dpi=200)
plt.show()

# =========================================================
# REPORT NOTE
# =========================================================
report_note = (
    "This notebook satisfies the requirement by generating 15 image pairs from the user's own face, "
    "where each pair consists of the aligned original face and one perturbed variant. "
    "Attack success rate is reported as the fraction of perturbations whose verifier cosine similarity "
    "to the original drops below a calibrated threshold. Visual similarity is reported using SSIM, PSNR, and LPIPS."
)
note_path = tables_dir / "personal_attack_report_note.txt"
note_path.write_text(report_note, encoding="utf-8")

print("\nSaved:")
print("Aligned original :", aligned_save_path)
print("Pairs CSV        :", pairs_csv)
print("Summary JSON     :", summary_path)
print("Overview grid    :", overview_grid)
print("Similarity plot  :", sim_plot)
print("SSIM plot        :", ssim_plot)
print("LPIPS plot       :", lpips_plot)
print("\nNotebook 10 complete.")
```
